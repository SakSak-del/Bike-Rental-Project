Meanwhile, models like **Random Forest** and **XGBoost Regressor** also performed very well with realistic and slightly higher error metrics, making them suitable candidates for real-world predictions where a bit of generalization is required.

üìå **Recommendation:**

```python
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)

print("Shape of X_train: ",X_train.shape)
print("Shape of y_train: ",y_train.shape)
print("Shape of X_test: ",X_test.shape)
print("Shape of y_test: ",y_test.shape)
```

## Feature Scaling using Standard Scaler

```python
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

## Model Building

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
lr_model.score(X_test,y_test)
```

### Ridge Regression (L2 Regularization)


```python
from sklearn.linear_model import Ridge

r_model = Ridge()
r_model.fit(X_train,y_train)
r_model.score(X_test,y_test)
```

### Lasso Regression (L1 Regression)

```python
from sklearn.linear_model import Lasso

l_model = Lasso()
l_model.fit(X_train,y_train)
l_model.score(X_test,y_test)
```

### DecisionTree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(max_depth=5) 
dt_model.fit(X_train, y_train)
dt_model.score(X_test,y_test)
```

### RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
score = rf_model.score(X_test, y_test)
print("Random Forest R^2 Score:", score)
print(rf_model) 
```

### XGBRegressor

```python
from xgboost import XGBRegressor
xg_model = XGBRegressor()
xg_model.fit(X_train,y_train)
xg_model.score(X_test,y_test)
```

### Determining the suitable Model

```python
models = {
    'Linear Regression': l_model,
    'Ridge Regression': r_model,
    'Lasso Regression': lr_model,
    'DecisionTree Regressor' : dt_model,
    'Random Forest': rf_model,
    'XGBRegressor' : xg_model
}
results={}
from sklearn.metrics import root_mean_squared_error, r2_score , mean_absolute_error
for name,model in models.items():
    y_pred = model.predict(X_test)
    rmse = root_mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    MAE = mean_absolute_error(y_test, y_pred)
    model_score = model.score(X_test,y_test)
    results[name] = {'RMSE': rmse, 'R¬≤': r2, 'MAE': MAE,'Model Score' : model_score}

results_df = pd.DataFrame(results).T
results_df
```

From the table, the **Linear Regression** and **Lasso Regression** models both achieve a perfect **R¬≤ score** of **1.0000** and exhibit extremely low error metrics (**RMSE** and **MAE** close to zero). 
These results suggest these models fit the **data exceptionally well** and may be **overfitting** if the dataset is small or lacks diversity.
